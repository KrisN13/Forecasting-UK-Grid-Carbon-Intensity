# Converting the Dates and Resampling Time to Hourly

```python
df_carbon_raw = df_carbon_raw.sort_values('DATETIME')

# Convert the DATETIME column to datetime type
df_carbon_raw['DATETIME'] = pd.to_datetime(df_carbon_raw['DATETIME'])

# Set the DATETIME column as the index
df_carbon_raw = df_carbon_raw.set_index('DATETIME')

# Now you can resample
dfcarbon_resampled = df_carbon_raw.resample('h').mean()

# If you want to keep DATETIME as a column instead of an index after resampling
df_carbon_raw = dfcarbon_resampled.reset_index()
```

# Add Day of the Week Column

```python
df_carbon_raw['DATETIME'] = pd.to_datetime(df_carbon_raw['DATETIME'], utc=True)
df_carbon_raw = df_carbon_raw.sort_values('DATETIME')
df_carbon_raw = df_carbon_raw.set_index('DATETIME')
df_carbon_raw = df_carbon_raw.asfreq('h')

print(df_carbon_raw.info())
```

```python
expected_range = pd.date_range(
    start=df_carbon_raw.index.min(),
    end=df_carbon_raw.index.max(),
    freq='h'
)

missing_timestamps = expected_range.difference(df_carbon_raw.index)
missing_timestamps[:20]
len(missing_timestamps)
```
